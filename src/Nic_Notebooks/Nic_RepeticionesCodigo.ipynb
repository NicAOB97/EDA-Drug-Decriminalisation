### DIBUJAR TODAS LAS GRAFICAS DE CAMBIOS EN MUERTES EN TIEMPO

```python
drug_deaths_years_complete = pd.read_csv('../Nic_Data/Nic_CreatedData/drug_deaths_years_complete.csv')
```

```python
drug_deaths_years_complete = drug_deaths_years_complete.transpose()
```

```python
sns.set_theme()
sns.set(rc={'axes.facecolor':'#A997DF', 'figure.facecolor':'#A997DF', 'figure.figsize':(10,10) })

plt.xticks(rotation=55)
g = sns.lineplot(x=drug_deaths_years_complete.index, y=drug_deaths_years_complete['Germany'],
             data=drug_deaths_years_complete, color='#1A3A3A', linewidth=2.5)

g.axvline(8, color='black')
```
